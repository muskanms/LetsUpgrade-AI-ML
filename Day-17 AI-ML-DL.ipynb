```python
import pandas as pd
import seaborn as sns
import numpy as np
import os

```

```python
sns.boxplot(x = data["hp"], data = data)
```

```python
Mean = np.mean(data["hp"])

SD = np.std(data["hp"])
```

```python
print(Mean,"is the mean")
print(SD,"is the standard deviation")
```

```python
def outlier_checker(data):
    threshold = 2
    outlier = []
    for i in data:
        z_score = (i - Mean)/SD
        if z_score > threshold:
            outlier.append(i)
    print(outlier)
outlier_checker(data["hp"])
```
